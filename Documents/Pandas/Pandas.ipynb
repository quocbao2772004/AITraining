```python
import requests
from bs4 import BeautifulSoup as bts

url = 'https://vnexpress.net/nhung-dai-hoc-dau-tien-cong-bo-diem-chuan-nam-2022-4511498.html';
res = requests.get(url);
# res => response: phản hồi
#in thử ra xem status_code của nó ra bao nhiêu ?
print(res)

```

```python
data = []
tup = ()
if res.status_code == 200:
  soup = bts(res.content, 'html.parser')
  table = soup.find('tbody')
  if table:
    rows = table.find_all('tr')
    if rows:
      for i in range(1, len(rows)):
        item = rows[i].find_all('td')
        pos = item[0].text.strip()
        name = item[1].text.strip()
        range_point = item[2].text.strip().replace('\n', ' ')
        tup = (pos, name, range_point)
        data.append(tup)
for item in data:
  print(f"{item[0]} \t {item[1]} \t {item[2]}")
```

🫥*df.groupby(): Nhóm dữ liệu theo một cột hoặc nhiều cột.*

```python
df.drop(0, axis=0) #theo chieu ngang
df.drop('name', axis=1) #theo chieu doc
```

```python
data = {'name': ['toan', 'lam', 'hai', 'thai anh'],
        'sex': ['nam', 'nam', 'nam', 'nam'],
        'age': [18, 20, 15, 21],
        'hometown': ['xuan mai', 'ha noi', 'ha noi', 'ha noi'],
        'university': ['ptit', 'ptit', 'ptit', 'epu']}

df = pd.DataFrame(data)
df.rename(columns={"age": "years old"}, inplace=True)
print(df)

df = df.sort_values(by='years old')
print(df)
```

```python
grouped_df = df.groupby(['sex', 'name']).all()
grouped_df
```

**Tính Toán và Thống Kê:**<br>

*df.mean(), df.median(): Tính trung bình và trung vị.*<br>
*df.std(), df.var(): Tính độ lệch chuẩn và phương sai.*<br>
*df.sum(), df.min(), df.max(): Tính tổng, giá trị nhỏ nhất và lớn nhất.*

```python
df['years old'].mean()
df['years old'].median()
df['years old'].sum()
```

*Hàm lọc* : df.loc[]

```python
df.loc[df['years old'] > 18 ]
df.loc[(df['years old'] >18) | (df['university'] == 'ptit')]
```
