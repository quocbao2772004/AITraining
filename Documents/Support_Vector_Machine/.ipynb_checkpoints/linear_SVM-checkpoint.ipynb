# Support Vector Machines

## - Intro

### Support Vector Machines (SVM) was proposed by Vapnik and his colleages in 1970s. Then it became famous and popular in 1990s.
### Originally, SVM is a method for linear classification. It finds a hyperplane to separate the two classes of data.
### For non-linear classification for which no hyperplane separates well the data, kernel functions will be used. Kernel functions play the role to transform the data into another space, in which the data is linearly separable.

## - Advantages

* ### SVM has a strong theory that supports its performance.
* ### It can work well with very high dimensional problems.
* ### It is now one of the most popular and strong methods.
* ### For text categorization, SVM performs very well.

## - SVM: the linearly separable case

### Training data D = {$(\textbf{x}_1, y_1), (\textbf{x}_2, y_2), …, (\textbf{x}_r, y_r)$} with r instances.
### Each $\textbf{x}_i$ is a vector in an n-dimensional space, $\textbf{x}_i = (x_{i_1}, x_{i_2}, …, x_{i_n})^T$.
### $y_i$ is a class label in {-1; 1}. ‘1’ is possitive class, ‘-1’ is negative class.
### Linear separability assumption: there exists a hyperplane (of linear form) that well separates the two classes

```python
from IPython import display
display.Image("./Img/svm1.PNG")
```

### SVM finds a hyperplane of the form: $$f(x) = \langle\textbf{w} . \textbf{x}\rangle + b$$
### **w** is the weight vector; b is a real number (bias).
### Such that for each $\textbf{x}_i$: $$y_i = \begin{cases}1&if&\langle \textbf{w}. \textbf{x}_i\rangle + b \ge 0 \\-1&if& \langle \textbf{w}.\textbf{x}_i\rangle + b < 0\end{cases}$$
### The hyperplane ($H_0$) which separates the possitive from negative class is of the form: $$\langle\textbf{w} . \textbf{x}\rangle + b = 0$$

```python
display.Image("./Img/svm2.PNG")
```

### But there might be infinitely many separating hyperplanes. Which one should we choose?

### Answer: SVM selects the hyperplane with **max margin**
### It is proven that the max-margin hyperplane has minimal errors among all possible hyperplanes.

## Margin?

### Denote (x+,1) in possitive class and ($x^-$,-1) in negative class which are closest to the separating hyperplane $H_0 \space (\langle\textbf{w}.\textbf{x}\rangle + b = 0)$
### Define two parallel marginal hyperplanes as follows:
* #### $H_+$ crosses $\textbf{x}^+$ and is parallel with $H_0$: $\langle\textbf{w}.\textbf{x}^+\rangle + b = 1$
* #### $H_+$ crosses $\textbf{x}^-$ and is parallel with $H_0$: $\langle\textbf{w}.\textbf{x}^-\rangle + b = 1$
* #### No data point lies between these two marginal hyperplanes. And satisfying:
$$\langle\textbf{w}.\textbf{x}_i\rangle + b \ge 1 \space if \space y_i = 1$$
$$\langle\textbf{w}.\textbf{x}_i\rangle + b \le -1 \space if \space y_i = -1$$

```python
display.Image("./Img/svm3.PNG")
```

### **Margin** is defined as the distance between the two marginal hyperplanes.
* #### Denote $d_+$ the distance from $H_0$ to $H_+$.
* #### Denote $d_-$ the distance from $H_0$ to $H_-$.
* #### ($d_+ + d_-$) is the margin.
### The distance from a point $\textbf{x}_i$ to the hyperplane $H_0$ ($\langle\textbf{w}.\textbf{x}\rangle + b = 0$) is computed as: $$\frac{|\langle\textbf{w}.\textbf{x}\rangle + b|}{\|\textbf{w}\|}$$
### Where: $$\|\textbf{w}\| = \sqrt{\langle\textbf{w}.\textbf{w}\rangle} = \sqrt{w_1^2 + w_2^2 + ... + w_n^2}$$

```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
from __future__ import print_function
import numpy as np 
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
np.random.seed(22)
```

```python
means = [[2, 2], [4, 2]]
cov = [[.3, .2], [.2, .3]]
N = 10
X0 = np.random.multivariate_normal(means[0], cov, N)
X1 = np.random.multivariate_normal(means[1], cov, N)
X = np.concatenate((X0.T, X1.T), axis = 1)
y = np.concatenate((np.ones((1, N)), -1*np.ones((1, N))), axis = 1)
```

```python
# plot points
plt.plot(X0[:, 0], X0[:, 1], 'bs', markersize = 8, alpha = .8)
plt.plot(X1[:, 0], X1[:, 1], 'ro', markersize = 8, alpha = .8)
plt.axis('equal')
# axis limits
plt.ylim(0, 3)
plt.xlim(2, 4)


# hide tikcs 
cur_axes = plt.gca()
cur_axes.axes.get_xaxis().set_ticks([])
cur_axes.axes.get_yaxis().set_ticks([])

plt.xlabel('$x_1$', fontsize = 20)
plt.ylabel('$x_2$', fontsize = 20)
plt.show()
```

```python
from sklearn.svm import SVC

clf = SVC(kernel = 'linear', C = 1e5) # just a big number 
y = y.reshape((2*N,))
clf.fit(X.T, y ) # each sample is one row

w = clf.coef_
b = clf.intercept_
print('w = ', w)
print('b = ', b)
```

```python
fig, ax = plt.subplots()

x1 = np.arange(-10, 10, 0.1)
y1 = -w.T[0, 0]/w.T[1, 0]*x1 - b/w.T[1, 0]
y2 = -w.T[0, 0]/w.T[1, 0]*x1 - (b-1)/w.T[1, 0]
y3 = -w.T[0, 0]/w.T[1, 0]*x1 - (b+1)/w.T[1, 0]
plt.plot(x1, y1, 'k', linewidth = 3)
plt.plot(x1, y2, 'k')
plt.plot(x1, y3, 'k')

y4 = 10*x1
plt.plot(x1, y1, 'k')

plt.plot(X0[:, 0], X0[:, 1], 'bs', markersize = 8, alpha = .8)
plt.plot(X1[:, 0], X1[:, 1], 'ro', markersize = 8, alpha = .8)

plt.axis('equal')
plt.ylim(0, 3)
plt.xlim(2, 4)

# hide tikcs 
cur_axes = plt.gca()
cur_axes.axes.get_xaxis().set_ticks([])
cur_axes.axes.get_yaxis().set_ticks([])

plt.xlabel('$x_1$', fontsize = 20)
plt.ylabel('$x_2$', fontsize = 20)

plt.show()
```
