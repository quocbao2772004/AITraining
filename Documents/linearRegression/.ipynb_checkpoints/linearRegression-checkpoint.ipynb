Linear Regression - Hồi quy tuyến tính( một thuật toán thuộc kiểu **supervised machine learning** (học có giám sát) )


**Hồi quy tuyến tính là gì?**
Hồi quy tuyến tính là một kỹ thuật phân tích dữ liệu dự đoán giá trị của dữ liệu không xác định bằng cách sử dụng một giá trị dữ liệu liên quan và đã biết khác.Có 2 loại đó là : **Simple Linear Regression** ( Mô hình này chỉ có một biến độc lập) và **Multiple Linear Regression**(Mô hình này có nhiều hơn một biến độc lập)

**Một số ví dụ :**
***Dự báo giá cả***: dự đoán giá nhà, giá cổ phiếu, giá nhiên liệu dựa trên các yếu tố như vị trí, kích thước, chất lượng, lượng cung cầu, …

***Dự báo điểm số:*** dự đoán điểm số của học sinh dựa trên thời gian học, nỗ lực, kỹ năng, trình độ giáo viên, …

***Dự báo sản phẩm:*** dự đoán đầu ra sản xuất dựa trên thời gian, công suất, nguyên liệu, lao động, …

**2, Phân tích toán học:**

---
Với  hồi quy tuyến tính thì ta có thể dùng toán học để dễ hình dung như sau : Ví dụ Giá của một căn nhà phụ thuộc vào độ rộng **x1**, số phòng **x2**, và khoảng cách với trung tâm thành phố **x3** thì ta có thể biểu diễn giá **y** của ngôi nhà đó như sau:

---

$$y \approx  f(\mathbf{x}) = \hat{y} $$
$$ f(\mathbf{x}) =w_1 x_1 + w_2 x_2 + w_3 x_3 + w_0 ~~~~ $$


ta có thể hiểu : **f(x)**, **x1**, **x2**, **x3** là mối quan hệ tuyến tính với
**f(x)** là biến đầu ra còn **x1**,**x2**,**x3** là các biến đầu vào.


*   $$w_1, w_2, w_3, w_0$$ là các trọng số. Và lời giải tìm được các hệ số tối ưu chính là lời giải của bài toán linear regression.

---
**Chú ý** :
*  $$\hat{y} $$  là giá trị của mô hình linear regression dự đoán dựa trên số liệu thống kê chúng ta có trong tập training data
* y là giá trị thực

---
Giữa 2 giá trị sẽ có sự khác nhau và ta muốn mô hình của chúng ta sẽ có sự sai số giữa 2 giá trị rất nhỏ.


**3,Phương pháp bình phương nhỏ nhất**: Least Squares Method
$$L(w) = Σᵢ(f(xᵢ) - yᵢ)²$$
Hàm $$L(w)$$ được gọi làm hàm mất mát (loss function)

Để dùng phương pháp bình phương nhỏ nhất hiệu quả thì thì ta đưa về ma trận như sau:

---



*   $$\mathbf{w} = [w_0, w_1, w_2, w_3]^T$$ là vector cột hệ số
*   $$\mathbf{\bar{x}} = [1, x_1, x_2, x_3]$$ là vector hàng dữ liệu đầu vào
* khi đó ta có thể viết lại:
$$y \approx \mathbf{\bar{x}}\mathbf{w} = \hat{y}$$

---
Hàm mất mát biểu diễn lại:  $$\mathcal{L}(\mathbf{w})
= \sum_{i=1}^N (y_i - \mathbf{\bar{x}}_i\mathbf{w})^2 =  \|\mathbf{y} - \mathbf{\bar{X}}\mathbf{w} \|_2^2$$

---






*   Với $$\mathbf{y} = [y_1; y_2; \dots; y_N]$$ là một vector cột chứa tất cả các output của training data
*   $$\mathbf{\bar{X}} = [\mathbf{\bar{x}}_1; \mathbf{\bar{x}}_2; \dots; \mathbf{\bar{x}}_N ]$$là ma trận dữ liệu đầu vào (mở rộng) mà mỗi hàng của nó là một điểm dữ liệu
---
Sau khi biến đổi thì ta cần tìm nghiệm phương trình đạo hàm bằng 0  của $$L(w)
$$ để tìm được $$minL(w)$$
---
Đạo hàm của hàm mất mát là:
$$\frac{\partial{\mathcal{L}(\mathbf{w})}}{\partial{\mathbf{w}}}
= \mathbf{\bar{X}}^T(\mathbf{\bar{X}}\mathbf{w} - \mathbf{y})$$






**4,Ngoại lệ**: phương trình đạo hàm có vô số nghiệm hoặc không có nghiệm.

**5,Hạn chế của hồi quy tuyến tính**: nhạy cảm với nhiễu

```python
# Import các thư viện cần thiết
!pip install sklearn matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

# Tạo dữ liệu mẫu
data = np.array([
    [15, 1504],
    [17, 1608],
    [75, 7394],
    [10, 831],
    [50, 5507],
    [35, 3011],
    [90, 8612],
    [80, 9036],
    [5, 614],
    [65, 5832]
])
area = data[:, 0]
price = data[:, 1]

# Vẽ biểu đồ dữ liệu
plt.scatter(area, price, color='red')
plt.xlabel("Diện tích (m2)")
plt.ylabel("Giá nhà (Triệu đồng)")

# Huấn luyện mô hình hồi quy tuyến tính
regr = linear_model.LinearRegression()
area_2D = area[:, np.newaxis]
regr.fit(area, price)

# Vẽ đường hồi quy
plt.plot(area, regr.predict(area_2D), color='blue')
plt.show()

# Dự đoán giá nhà
need_prediction = [19.2, 102.5, 56.9]
for element in need_prediction:
    print(f"Căn nhà có diện tích {element:.1f} m2 có giá khoảng {regr.predict([[element]])[0]:.2f} triệu đồng")
```
